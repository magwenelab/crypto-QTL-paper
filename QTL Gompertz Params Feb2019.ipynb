```python
## QTL analysis
## Bring in needed mods
import pandas as pd, numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use(['dark_background','tableau-colorblind10'])
```

```python
lp = '../FILES/Tecan_qtlruns_Dec172018_curated_baselined_gompertz_pvalues.csv'
```

```python
df = pd.read_csv(lp,index_col=0)
```

```python
df.head()
```

```python
chrlist = df.Chrom.unique()
clens = np.array([df[(df.Chrom==c)].Pos.max() for c in chrlist])
cumpos = [0] + list(np.cumsum(clens[:-1]))
```

```python
df_col = df.columns.tolist()
```

```python
cols_auc_mean = [a for a in df_col if 'AUC' in a.split('_') and 
                      'mean' in a.split('_')]
cols_auc_median = [a for a in df_col if 'AUC' in a.split('_') and 
                       'median' in a.split('_')]
```

```python
## Set mid points
chrom_midpts = 0.5 * np.sum(np.array([a 
                    for a in zip([0]+list(np.cumsum(clens)),
                            np.cumsum(clens))]),axis=1)
```

```python
cols_a_mean = [a for a in df_col if 'A' in a.split('_') and 
                      'mean' in a.split('_')]
cols_a_median = [a for a in df_col if 'A' in a.split('_') and 
                       'median' in a.split('_')]
```

```python
cols_l_mean = [a for a in df_col if 'l' in a.split('_') and 
                      'mean' in a.split('_')]
cols_l_median = [a for a in df_col if 'l' in a.split('_') and 
                       'median' in a.split('_')]
```

```python
cols_u_mean = [a for a in df_col if 'u' in a.split('_') and 
                      'mean' in a.split('_')]
cols_u_median = [a for a in df_col if 'u' in a.split('_') and 
                       'median' in a.split('_')]
```

```python
fig, ax = plt.subplots(len(cols_u_median),1,figsize=(20,25),
                       sharex=True,sharey=True)
for p,pheno in enumerate(cols_u_median):
    plt.sca(ax[p]);
    plt.title(pheno,x=0.1,y=0.7,fontsize=18)
    for c,ch in enumerate(chrlist):
        temp = df[(df.Chrom==ch)][['Pos',pheno]]
        plt.plot(temp.Pos.values+cumpos[c],temp[pheno].values,'.')
ax[-1].set_xticks(chrom_midpts);
ax[-1].set_xticklabels([str(i) for i in range(1, len(clens)+1)]);
```

```python
fig, ax = plt.subplots(len(cols_l_median),1,figsize=(20,25),
                       sharex=True,sharey=True)
for p,pheno in enumerate(cols_l_median):
    plt.sca(ax[p]);
    plt.title(pheno,x=0.1,y=0.7,fontsize=18)
    for c,ch in enumerate(chrlist):
        temp = df[(df.Chrom==ch)][['Pos',pheno]]
        plt.plot(temp.Pos.values+cumpos[c],temp[pheno].values,'.')
ax[-1].set_xticks(chrom_midpts);
ax[-1].set_xticklabels([str(i) for i in range(1, len(clens)+1)]);
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,10),
                       sharex=True,sharey=True)
for c,ch in enumerate(chrlist):
    temp = df[(df.Chrom==ch)][['Pos']+cols_u_median]
    plt.plot(temp.Pos.values+cumpos[c],temp[cols_u_median].max(axis=1).values,'.')
plt.gca().set_prop_cycle(None)
for c,ch in enumerate(chrlist):
    temp = df[(df.Chrom==ch)][['Pos']+cols_u_mean]
    plt.plot(temp.Pos.values+cumpos[c],-temp[cols_u_mean].max(axis=1).values,'.')
ax.set_xticks(chrom_midpts);
ax.set_xticklabels([str(i) for i in range(1, len(clens)+1)]);
```
